```python
import csv

# Prepare a list to store unique vehicle types
vehicle_types = []

# Open the file
with open('../data/test_data.csv', 'r') as file:
    # Create a CSV reader
    reader = csv.reader(file, delimiter=';')

    # Extract the headers
    headers = next(reader)

    # Prepare lists to store the normalized data
    trajectories = []
    positions = []

    # Loop through the remaining rows
    for row in reader:
        # Extract trajectory information
        track_id = row[0]
        vehicle_type = row[1].strip()  # Strip leading and trailing spaces

        # Check if the vehicle type is already in the list
        if vehicle_type not in vehicle_types:
            # If it's not, add it to the list
            vehicle_types.append(vehicle_type)

        # Use the index of the vehicle type in the list as the vehicle type ID
        vehicle_type_id = vehicle_types.index(vehicle_type)

        traveled_distance = row[2]
        avg_speed = row[3]

        # Store the trajectory information
        trajectories.append([track_id, vehicle_type_id, traveled_distance, avg_speed])

        # Loop through the remaining columns in steps of 6
        for i in range(4, len(row), 6):
            # Check if there are enough remaining columns
            if i + 5 < len(row):
                lat = row[i]
                lon = row[i+1]
                speed = row[i+2]
                lon_acc = row[i+3]
                lat_acc = row[i+4]
                time = row[i+5]
        
                # Store the position information
                positions.append([track_id, lat, lon, speed, lon_acc, lat_acc, time])

# Write the normalized data to new CSV files
with open('../data/normalized_trajectories.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Track ID', 'Vehicle Type ID', 'Traveled Distance', 'Average Speed'])
    writer.writerows(trajectories)

with open('../data/normalized_positions.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Track ID', 'Latitude', 'Longitude', 'Speed', 'Longitudinal Acceleration', 'Lateral Acceleration', 'Time'])
    writer.writerows(positions)

# Write the unique vehicle types to a new CSV file
with open('../data/vehicle_types.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Vehicle Type ID', 'Vehicle Type'])
    writer.writerows(enumerate(vehicle_types))
```

```python
import csv

# Prepare a list to store unique vehicle types
vehicle_types = []

# Open the file
with open('../data/test_data.csv', 'r') as file:
    # Create a CSV reader
    reader = csv.reader(file, delimiter=';')

    # Extract the headers
    headers = next(reader)

    # Prepare lists to store the normalized data
    trajectories = []
    positions = []

    # Loop through the remaining rows
    for row in reader:
        # Extract trajectory information
        track_id = row[0]
        vehicle_type = row[1].strip()  # Strip leading and trailing spaces

        # Check if the vehicle type is already in the list
        if vehicle_type not in vehicle_types:
            # If it's not, add it to the list
            vehicle_types.append(vehicle_type)

        # Use the index of the vehicle type in the list as the vehicle type ID
        vehicle_type_id = vehicle_types.index(vehicle_type)

        traveled_distance = row[2]
        avg_speed = row[3]

        # Store the trajectory information
        trajectories.append([track_id, vehicle_type_id, traveled_distance, avg_speed])

        # Loop through the remaining columns in steps of 2 for positions
        for i in range(4, len(row), 2):
            # Check if there are enough remaining columns
            if i + 1 < len(row):
                lat = row[i]
                lon = row[i+1]

                # Store the position information
                positions.append([track_id, lat, lon])

# Write the normalized data to new CSV files
with open('../data/normalized_trajectories.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Track ID', 'Vehicle Type ID', 'Traveled Distance', 'Average Speed'])
    writer.writerows(trajectories)

with open('../data/normalized_positions.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Track ID', 'Latitude', 'Longitude'])
    writer.writerows(positions)

# Write the unique vehicle types to a new CSV file
with open('../data/vehicle_types.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Vehicle Type ID', 'Vehicle Type'])
    writer.writerows(enumerate(vehicle_types))
```
